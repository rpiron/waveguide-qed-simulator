I would like to check the relevence of the following self consistent equation:

$$ e^{\lambda \Delta t} - 1 = \Delta t \alpha_0 (1 - \alpha_1)^{-1} $$

where $\alpha_0$ and $\alpha_1$ are evaluated at $\Lambda \sim \Delta t^{-1}$

```python
gamma_0 = np.pi
omega_0 = 10

def get_bare_param(omega_0, gamma_0, dt):


    omega_A = omega_0
    gamma_A = -2 / dt * np.log( np.abs(np.cos ( np.sqrt(gamma_0 * dt))))

    return omega_A, gamma_A

get_bare_param(omega_0, gamma_0, dt=1)
```

```python
dt_tab = np.linspace(0.1, 3, 200)
bare_params = np.array([get_bare_param(omega_0, gamma_0, dt) for dt in dt_tab])
plt.figure(figsize=(6, 4))
plt.plot(dt_tab, bare_params[:, 0], label=r"$\omega_A$")
plt.plot(dt_tab, bare_params[:, 1], label=r"$\gamma_A$")
plt.xlabel(r"$\Delta t$")
plt.legend()
plt.tight_layout()
plt.show()
```
