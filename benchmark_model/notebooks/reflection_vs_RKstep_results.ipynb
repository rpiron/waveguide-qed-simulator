<h1> Experiment reflection against number of RG steps </h1>

```python
#General imports
resol = 300
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "axes.linewidth": 0.7,
    "xtick.major.width": 0.6,
    "ytick.major.width": 0.6,
    "xtick.direction": "in",
    "ytick.direction": "in",

    # Lines / markers
    "lines.linewidth": 1.0,
    "lines.markersize": 4,

})

import sys
import numpy as np
from pathlib import Path
pi = np.pi

project_root = Path().resolve().parents[1]
sys.path.append(str(project_root))

#Local imports
from benchmark_model.scripts_experiments.reflection_vs_RKstep import run_reflection_vs_RKstep

# Function to compute the converged value of a function
def final_value(y, n_avg=20):
    return np.mean(y[-n_avg:])
```

<h2> Execute the experiment </h2>

Prepapre the experiment

```python
#Time evolution
T = 50  
nb_pts_steps = 50
N_step_tab = np.unique(np.logspace(1, 4, nb_pts_steps).astype(int))
dt_tab = T / N_step_tab

#Physical parameters
Omega_A = 10*pi
g_A = np.sqrt(pi/2)
param_atom = {'Omega_A': Omega_A, 'g_A': g_A, 'L': 100}

omega_p_tab = [10*pi, 9.75*pi, 9.5*pi]
cutoffs_tab = [(10*pi - k*pi, 10*pi + k*pi) for k in range(1,11)]

#incoming photon parameters
param_photon_bis = {'delta_k': 0.05*np.pi, 'x_0': -25} 
```

Run the loop

```python
for (omega_p, i) in zip(omega_p_tab, range(len(omega_p_tab))):
    for (cutoff, j) in zip(cutoffs_tab, range(len(cutoffs_tab))):
        index_omega_p, index_cutoff = i+1, j+1
        
        print(f"Running for omega_p = {omega_p:.2f} and cutoff = {cutoff}")

        param_photon = {'omega_p': omega_p, 'delta_k': 0.05*np.pi, 'x_0': -50/4} 
        cutoffs = {"ir_cutoff": cutoff[0], "uv_cutoff": cutoff[1]}

        _= run_reflection_vs_RKstep(param_photon, param_atom, T, dt_tab, cutoffs, 
                                    index_experiment=index_cutoff, index_omega_p=index_omega_p)
        
```

<h2> Plot the results </h2>

Recover the data of a single frequency window

```python
index_experiment = 3

data_to_plot = []

for i in range(len(omega_p_tab)):
    file_data = f"../results/csv_files/reflection_vs_RKstep/reflection_vs_RKstep_omegap_{i+1}_xp{index_experiment}.csv"
    df = pd.read_csv(file_data)
    data_to_plot.append(df)

#Prepare the three curves corresponding to the three photonic frequencies
T = 50  
nb_pts_steps = 50
N_step_tab = np.unique(np.logspace(1, 4, nb_pts_steps).astype(int))
dt_tab = T / N_step_tab

#Get the time step required to reach convergence for each curve
conv_index_tab = np.array([np.where(np.abs(data_to_plot[0]['final_reflection'] - data_to_plot[0]['final_reflection'].iloc[-1]) < 0.05 * data_to_plot[0]['final_reflection'].iloc[-1])[0][0],
                  np.where(np.abs(data_to_plot[1]['final_reflection'] - data_to_plot[1]['final_reflection'].iloc[-1]) < 0.05 * data_to_plot[1]['final_reflection'].iloc[-1])[0][0],
                  np.where(np.abs(data_to_plot[2]['final_reflection'] - data_to_plot[2]['final_reflection'].iloc[-1]) < 0.05 * data_to_plot[2]['final_reflection'].iloc[-1])[0][0]])

conv_index = np.max(conv_index_tab)
dt_convergence = dt_tab[conv_index]
N_step_convergence = N_step_tab[conv_index]
```

Plot the results

```python
colors = ['#081f5c', '#a73c51', '#bda546']
labels = [
    r'$\omega_p = 10\pi$',
    r'$\omega_p = 9.75\pi$',
    r'$\omega_p = 9.5\pi$'
]
label_y_pos = [0.91, 0.72, 0.45]

fig, ax = plt.subplots(figsize=(3, 3), dpi=300)

#Plot the three curves
for i, (c, lab) in enumerate(zip(colors, labels)):
    
    #curve
    y = data_to_plot[i]['final_reflection'].values
    ax.plot(N_step_tab, y, '-', color=c, lw=1.0, label=lab, zorder=3)

    # 5% band around the converged value
    y_inf = final_value(y)
    y_low  = 0.95 * y_inf
    y_high = 1.05 * y_inf
    ax.fill_between(N_step_tab,y_low,y_high,color=c,alpha=0.12,linewidth=0, zorder=1)

# Vertical line at convergence point
ax.axvline(N_step_convergence,color='black',ls=':',lw=1.2, zorder=2)
ax.text(
    N_step_convergence * 1.1,
    0.1,
    f'$N_{{\\rm conv}} = {int(N_step_convergence)}$',
    fontsize=10,
    color='black'
)

# Labels and limits
ax.set_xlabel(r'\textbf{Number of RK4 steps} $N_{\rm step}$', fontsize=10)
ax.set_ylabel(r'\textbf{Reflection probability}', fontsize=10)
ax.set_xscale("log")
ax.set_xlim(1e1,1e4)
ax.set_xticks([1e1,1e2,1e3,1e4])
ax.set_ylim(0,1.1)
ax.grid(color='0.9', linestyle='-', linewidth=0.4)
ax.tick_params(axis='both', which='major', length=4, width=0.8, labelsize=10)

# Indicate frequency of the photons
for y_pos, lab, col in zip(label_y_pos, labels, colors):

    ax.text(
        0.5,                 
        y_pos,                
        lab,
        color=col,
        fontsize=10,
        va='center',
        ha='left',
        transform=ax.get_yaxis_transform()
    )

#font size
for item in [ax.xaxis.label, ax.yaxis.label]:
    item.set_fontsize(10)

for item in (ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(10)

plt.tight_layout()
plt.savefig(f"../results/fig/experiment_convergence_RKstep_xp{index_experiment}.pdf", dpi=resol)
plt.show()
```

Overall behavior of the minimal number of steps for convergenve against bandwidth

```python
N_step_convergence_several_curves = []
bandwidth_tab = (np.array(cutoffs_tab)[:,1] - np.array(cutoffs_tab)[:,0])/2

for index_omega_p in range(1, len(omega_p_tab)+1):

    N_step_convergence_tab = []

    T = 50  
    nb_pts_steps = 50
    N_step_tab = np.unique(np.logspace(1, 4, nb_pts_steps).astype(int))
    dt_tab = T / N_step_tab

    for i in range(len(cutoffs_tab)):
        file_data = f"../results/csv_files/reflection_vs_RKstep/reflection_vs_RKstep_omegap_{index_omega_p}_xp{i+1}.csv"
        df = pd.read_csv(file_data)
        #Recover the minimal step required to reach convergence for this curve
        conv_index = np.where(np.abs(df['final_reflection'] - df['final_reflection'].iloc[-1]) < 0.05 * df['final_reflection'].iloc[-1])[0][0]
        dt_convergence = df['dt'].values[conv_index]

        N_step_convergence = int(T / dt_convergence)
        N_step_convergence_tab.append(N_step_convergence)
    
    N_step_convergence_several_curves.append(N_step_convergence_tab)
```

```python
colors = ['#081f5c', '#a73c51', '#bda546']
labels = [
    r'$\omega_p = 10\pi$',
    r'$\omega_p = 9.75\pi$',
    r'$\omega_p = 9.5\pi$'
]

fig, ax = plt.subplots(figsize=(3, 3), dpi=300)

# Reference guides for N_conv proportional to the simulated bandwidth.
scaling_factors = [8, 12, 16, 20]
bandwidth_guide = np.linspace(bandwidth_tab.min(), bandwidth_tab.max(), 300)
n_conv_values = np.asarray(N_step_convergence_several_curves, dtype=float)

for f in scaling_factors:
    y_guide = f * bandwidth_guide
    ax.plot(
        bandwidth_guide,
        y_guide,
        '--',
        color='0.55',
        alpha=0.55,
        linewidth=0.7,
        zorder=1
    )
    ax.text(
        bandwidth_guide[-1] * 1.05,
        y_guide[-1],
        rf'${f}\Lambda$',
        color='0.4',
        fontsize=8,
        va='center',
        ha='left'
    )

for i, (c, lab) in enumerate(zip(colors, labels)):
    N_step_convergence_tab = N_step_convergence_several_curves[i]
    ax.plot(
        bandwidth_tab,
        N_step_convergence_tab,
        'o-',
        color=c,
        markerfacecolor='white',
        markeredgewidth=0.8,
        markersize=3.2,
        linewidth=1.0,
        label=lab,
        zorder=3
    )

#ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlim(bandwidth_tab.min() * 0.9, bandwidth_tab.max() * 1.2)
ax.set_ylim(0.9*1e1, 1.1*1e3)
#ax.set_ylim(
#    min(n_conv_values.min(), scaling_factors[0] * bandwidth_tab.min()) * 0.75,
#    max(n_conv_values.max(), scaling_factors[-1] * bandwidth_tab.max()) * 1.25)

ax.set_xticks([2*pi, 4*pi, 6*pi, 8*pi, 10*pi])

def pi_formatter(x, pos):
    n = x / np.pi
    if np.isclose(n, round(n)):
        return rf'${int(round(n))}\pi$'
    return rf'${n:.1f}\pi$'

ax.xaxis.set_major_formatter(FuncFormatter(pi_formatter))

ax.set_xlabel(r'\textbf{Bandwidth} $\Lambda$', fontsize=10)
ax.set_ylabel(r'\textbf{Convergence RK4 steps} $N_{\rm conv}$', fontsize=10)
ax.grid(color='0.9', linestyle='-', linewidth=0.4, which='major')
ax.grid(color='0.94', linestyle='-', linewidth=0.25, which='minor')
ax.tick_params(axis='both', which='major', length=4, width=0.8, labelsize=10)
ax.tick_params(axis='both', which='minor', length=2, width=0.5)
ax.legend(prop={'size': 9}, loc='lower right', frameon=False, handlelength=1.6)

plt.tight_layout()
plt.savefig(f"../results/fig/Nconv_vs_bandwidth.pdf", dpi=resol)
plt.show()
```
